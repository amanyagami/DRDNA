```python
import numpy as np

def calculate_emd(hist1, hist2):
    """
    Calculate the Earth Mover's Distance (EMD) between two histograms.
    
    Parameters:
    - hist1: First histogram (list or numpy array of frequencies)
    - hist2: Second histogram (list or numpy array of frequencies)
    
    Returns:
    - emd: The normalized Earth Mover's Distance between hist1 and hist2
    """
    
    # Convert histograms to numpy arrays if they aren't already
    hist1 = np.array(hist1, dtype=float)
    hist2 = np.array(hist2, dtype=float)
    
    # Compute the cumulative distributions for both histograms
    cdf1 = np.cumsum(hist1)
    cdf2 = np.cumsum(hist2)
    
    # Calculate the absolute differences between the cumulative distributions
    cumulative_diff = np.abs(cdf1 - cdf2)
    
    # Compute the total amount of work (sum of differences)
    emd_value = np.sum(cumulative_diff)
    
    # Normalize by the total number of data points in the first histogram
    total_freq_hist1 = np.sum(hist1)
    
    # Return the normalized EMD
    emd_normalized = emd_value / total_freq_hist1
    return emd_value

# Example usage:

# Define two histograms (these should be the frequencies in each bin)
hist1 = [2, 3, 5, 8]  # Example frequencies in histogram 1
hist2 = [3, 4, 6, 7]  # Example frequencies in histogram 2

# Calculate the EMD between the two histograms
emd_result = calculate_emd(hist1, hist2)

print(f"Normalized Earth Mover's Distance (EMD): {emd_result:.4f}")
```

```python
import numpy as np

def calculate_emd(hist1, hist2):
    # Extract sorted bin ranges from both histograms
    all_bins = sorted(set(hist1.keys()).union(hist2.keys()))
    
    # Normalize the histograms
    total1 = sum(hist1.values())
    total2 = sum(hist2.values())
    
    # Cumulative distributions
    cdf1, cdf2 = [], []
    cumulative_sum1, cumulative_sum2 = 0, 0
    
    for bin_range in all_bins:
        # Get the values for the current bin (default to 0 if not present)
        bin_value1 = hist1.get(bin_range, 0) / total1
        bin_value2 = hist2.get(bin_range, 0) / total2
        
        # Update cumulative sums
        cumulative_sum1 += bin_value1
        cumulative_sum2 += bin_value2
        
        # Append the CDF values
        cdf1.append(cumulative_sum1)
        cdf2.append(cumulative_sum2)
    
    # Calculate the EMD (sum of absolute differences between CDFs)
    emd = np.sum(np.abs(np.array(cdf1) - np.array(cdf2)))
    
    return emd

# Example histograms (dictionary with ranges as keys)
histogram1 = {(0, 1): 0.1, (1, 2): 0.3, (2, 3): 0.4, (3, 4): 0.2}
histogram2 = {(0, 1): 0.2, (1, 2): 0.2, (2, 3): 0.3, (3, 4): 0.3}

emd_value = calculate_emd(histogram1, histogram2)
emd_value
```
